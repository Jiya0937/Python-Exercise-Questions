```python
# WAP to create a class Student and display details

class Student:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        
    def display(self):
        print("Name    : ", self.name)
        print("Balance : ",self.balance)

s1 = Student("jiya", 50000)
s1.display()

    
```

```python
# 6. WAP to find largest of two numbers using class

class Largest:
    def find(self, a, b, c):
        if(a < b and b < c):
            print("Largest Number is C : ", c)
            
        elif(a < b and c < b):
            print("Largest Number is B : ", b)
        else:
            print("Largest Number is A : ", a)

L1 = Largest()
L1.find(2876, 78, 212)
```

```python
# 9. WAP to calculate simple interest using class

class SI:
    def __init__ (self, p, r, t):
        self.p = p
        self.r = r
        self.t = t

    def Display(self):
        print("Principal : ", self.p)
        print("Rate : ", self.r)
        print("Time : ", self.t)

        print("Simple Interest : ", (self.p*self.r*self.t)/100)

si = SI(1000, 2, 1)
si.Display()
```
