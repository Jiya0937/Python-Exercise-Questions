```python
# 6. WAP to find largest of two numbers using class

class Largest:
    def find(self, a, b, c):
        if(a < b and b < c):
            print("Largest Number is C : ", c)
            
        elif(a < b and c < b):
            print("Largest Number is B : ", b)
        else:
            print("Largest Number is A : ", a)

L1 = Largest()
L1.find(2876, 78, 212)
```

```python
# 9. WAP to calculate simple interest using class

class SI:
    def __init__ (self, p, r, t):
        self.p = p
        self.r = r
        self.t = t

    def Display(self):
        print("Principal : ", self.p)
        print("Rate : ", self.r)
        print("Time : ", self.t)

        print("Simple Interest : ", (self.p*self.r*self.t)/100)

si = SI(1000, 2, 1)
si.Display()
```

```python
# Create a class to calculate the average marks of a student.

class Student:
    def __init__(self, hindi, maths, english, science):
        self.hindi = hindi
        self.maths = maths
        self.english = english
        self.science = science

    def Display(self):
        print("Hindi      : ",self.hindi)
        print("Maths      : ",self.maths)
        print("English    : ", self.english)
        print("Science    : ", self.science)
        print("Persantage : ", (self.hindi + self.maths + self.english + self.science)/4)

s1 = Student(56, 78, 98, 90)
s1.Display()
```

```python
# Create a class to calculate bonus based on salary.
class Bonus:
    def __init__ (self, salary, persantage):
        self.salary = salary
        self.persantage = persantage
    def calculate(self):
        Bonus = (self.salary*self.persantage)//100
        print("Salary     : ", self.salary)
        print("persantage : ",self.persantage)
        print("Bonus      : ",Bonus)

b1 = Bonus(15000, 10)
b1.calculate()
        
```

```python
# Create a class to check account balance
class Account:
    def __init__ (self, Username, Amount, account_Number):
        self.name = Username
        self.amount = Amount
        self.no = account_Number
    def show(self):
        print("User name      : ",self.name)
        print("Balance        : ", self.amount)
        print("Account Number : ",self.no)

A1 = Account("Sandhya Vishwakarma", 72562, 75629873462)
A1.show()
```

```python
# WAP to create a class Student with instance variable name and class variable count to count total students.

class Student:

    count = 0
    def __init__(self, name, branch):
        self.name = name
        self.branch = branch
        
    def show(self,count):
        print("Name   : ", self.name)
        print("Branch : ",self.branch)
        print("Count  : ", count+1)

count = 0
s1 = Student("Jiya Deshmukh", "AIML")
s1.show(count)
count +=1
s2 = Student("Sandhya Vishwakarma" , "AIML")
s2.show(count)
        
```
