```python
# WAP to swap two numbers.

a = int(input("Enter A and B : "))
b = int(input("Enter A and B : "))
print(a,b)
temp = a
a = b
b = temp
print(a,b)
```

```python
# WAP to take multiple inputs in one line.
# for string type
a, b, c = input("Enter the value of a and b and c : ").split()

print(" I      : ", a)
print(" Sister : ",b)
print(" Friend : ",c)

# for integer type
x, y, z = map(int, input("Enter Values : ").split())
print("Sum : ", x*y*z)
```
