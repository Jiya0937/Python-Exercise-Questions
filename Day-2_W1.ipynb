```python
# *2. WAP to find all pairs in a list whose sum is equal to a given number*

class find_Target:
    def target (self, arr, target):
        n = len(arr)
        found = False
        for i in range (n):
            for j in range(i+1, n):
                if(arr[i] + arr[j] == target):
                    print(i, j)
                    found = True
        if not found:
            return [-1, -1]
        
      

l = [3, 52, 24, 2, 22, 242, 7, -2]
t = find_Target()
t.target(l, 4)
```

```python
# 3. Shift elements to the right and move last elements to the front.
ele = [2, 4, 6, 8, 10] # output should be if k = 2 ele = [6, 8, 10, 2, 4]

def rotation(element, k):
    rotate_list = []
    remaining_ele = []
    for i in range(len(element)):
        if (i < k):
            rotate_list.append(element[i])
        else:
            remaining_ele.append(element[i])
    new_list = remaining_ele + rotate_list
    return new_list
    
ele = [2, 4, 6, 8, 10] # output should be if k = 2 ele = [6, 8, 10, 2, 4]
rotation(ele, 2)

# by the slicing
def rotate(ele, k):
    rotate_list = ele[k:] + ele[:k]
    return rotate_list
element = [3, 4, 6, 6, 7, 33, 75, 35]
rotate(element, 3)
```

```python
# 4. WAP to remove all occurrences of a given element from a list*

def remove_Occurance(arr, k):
    result = []
    for i in arr:
        if i != k:
            result.append(i)
    print(result)

l = [1, 3, 4, 3, 2]
remove_Occurance(l, 3)
```

```python
# 5.  *WAP to find common elements between two lists*
l1 = [2, 44, 4, 6, 46, 4,  1]
l2 = [9, 8, 7, 46, 3, 2, 4]
def find(lst1, lst2):
    common = []
    for i in lst1:
        if(i in lst2 and i not in common):
            common.append(i)
    print(common)
find(l1, l2)
```

```python
# 6. WAP to flatten a nested list (1 level deep

def flatten(lst):
    result = []
    for i in lst:
        if isinstance(i, list):
            result.extend(flatten(i))
        else:
            result.append(i)
    return result

l = [3, 23,[3, 5], [7, 89, 67, [78,89, 67, 47, 13]]]
flatten(l)
```

```python
# 7. WAP to find frequency of each element in a list*
l = ["Jiya", "sandhya", "Janhavi", "Drashti", "Lovely", "Assessment"]
def frequency(name):
    freq = {}
    count = 1
    for i in range(len(name)):
        if name[i] in freq:
            freq[name[i]] += 1
        else:
            freq[name[i]] = count
    print(freq)
        
frequency(l)
```

```python
# 8. WAP to find the longest increasing sublist

def longest_sublist(arr):
    max_list = []
    current = [arr[0]]
    for i in range(1, len(arr)):
        if (arr[i] > arr[i-1]):
            current.append(arr[i])
        else:
            if(len(current) > len(max_list)):
                max_list = current
            current = [arr[i]]
    if (len(current) > len(max_list)):
        max_list = current
    return max_list

arr = [3, 23, 3, 5, 7, 89, 67, 78, 89, 67, 47, 13]
longest_sublist(arr)
                
```

```python
# 8. WAP to find the longest increasing sublist
l = [3, -4, 2, 11, -1]
def max_subArray(arr):
    curr_sum = arr[0]
    max_sum = 0
    for i in range(1, len(arr)):
        curr_sum = curr_sum + arr[i]
        if(curr_sum < 0):
            curr_sum = 0
        max_sum = max(max_sum,curr_sum)
    print("Maximum_Subarray : ", max_sum)
max_subArray(l)   
```

```python
# 10. WAP to find duplicate elements in a list*
class find_duplicate:
    def found(self, nums):
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if(nums[i] == nums[j]):
                    print(nums[j])

l = [2, 2, 4, 3, 42, 23, 4]
d1 = find_duplicate()
d1.found(l)
```
